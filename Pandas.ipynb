```python
# To take average of data set . ==> donot forget work in column because axis = 0
col = data2.mean( axis = 0 )

col.head()
```

```python
# To take average of data set . ==> donot forget work in row because axis = 1
row = data2.mean( axis = 1)

row.head()
```

```python
# if data set save the shape of dictionary .
w = pd.Series({'a': 1 ,'b': 2 ,'c': 3 ,'d': 4 ,'e': 5})
x = pd.Series({'a': 6 ,'b': 7 ,'c': 8 ,'d': 9 ,'e': 10})
y = pd.Series({'a': 11 ,'b': 12 ,'c': 13 ,'d': 14 ,'e': 15})
z = pd.Series({'a': 16 ,'b': 17 ,'c': 18 ,'d': 19 ,'e': 20})

grades = pd.DataFrame({'Math':w,'Physics':x,'French':y,'Chemistry':z})
t_grades = grades.T

grades.head()
```

```python
# The trasnpose of data set .
t_grades.head()
```

```python
# To make sure about found keys or values in data set . key ==> key of dictionary
print('Math' in grades.keys())
print('math' in grades.keys())  # Python is make sensetive.
print(7 in grades.values)
print(50 in grades.values)
```

```python
# To sort specific column. sort ==> use sort_values()
sort = grades.sort_values(['Math'], ascending = False) # By default arrange ascending

sort.head()
```

```python
# To describe the data set that you work on it.
grades.describe()
```
